Before running this, inside Google Drive, create a folder named "mokuro-ocr" and create a folder for each individual manga **series** that will be ocr'ed. Then place the individual manga **volumes** inside its respective manga series folder. The directory should look like one of the two examples below:

```
MyDrive/
├─mokuro-ocr/
  ├─manga/
    ├─vol1/
```
or

```
MyDrive/
├─mokuro-ocr/
  ├─manga/
    ├─vol1.<zip/cbz>
```

For the commands below, replace ```<manga>``` with your manga series, ```<vol1>``` with the individual volume you want to OCR and ```<zip/cbz>``` with either "zip" or "cbz".

To run on individual volume folders:

To run on individual volume zip/cbz files:

````python
!mokuro "/content/drive/MyDrive/mokuro-ocr/<manga>/<vol1>.<zip/cbz>"
````

Run on a directory containing multiple volumes (either zip/cbz or folder format):

````python
!mokuro --parent_dir "/content/drive/MyDrive/mokuro-ocr/<manga>"
````
